## <u>Data Cleaning (cf. Module 10)</u>
##### **Describe any data cleaning methods applied (e.g., missing values, outliers, syntactic or semantic cleaning)**

The data pulled from Yahoo finance has a multi-level schema. Level 1 of the schema has the stock price attribute (i.e. opening price, closing price, adjusted closing price, volume etc. on a given day) and level 2 of the schema has the ticker symbol of the stock (NVDA, ABT, TSLA etc.). Below we have created a function called `data_formatter` that uses pd.MultiIndex in pandas to format the schema and retrieve only the adjusted closing price for each stock which is what we will use in our analysis.
Additionally, we have specified a datetime index for our dataframe to facilitate data integration.

Here is the code that we used to clean the reformat the data.

```python
def data_formatter(df):
    if isinstance(df.columns,pd.MultiIndex):
        lvl0=df.columns.get_level_values(0)
        if "Adj_Close" in set(lvl0):
            price=df["Adj_Close"].copy()
        
        else:
            raise KeyError("adj close price column not found")
    else:
        if 'Adj_Close' in df.columns:
            price=df[['Adj_Close']].rename(columns={'Adj Close':tickers[0]})
        else:
            raise KeyError('adj close price column not found')
    
    return price
```

```python
df_yf_cleaned=data_formatter(df_yf)
df_yf_cleaned
date_index = pd.date_range(start='2016-01-01', periods=len(df_yf_cleaned), freq='MS')
df_yf_cleaned.index=date_index
df_yf_cleaned.to_csv('stocks_cleaned_df_2016_2025',index=False)
```

## Data integration (cf. Module 7-8)
##### **Integration of datasets**

We will be using the `pandas` library in Python to integrate our datasets.  
We decided to perform an inner merge, and then after that we droped the NaN values so the dataset would have no missing values.
